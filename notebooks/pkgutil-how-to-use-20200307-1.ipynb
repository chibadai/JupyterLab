```python
module_list = []
for module_name in pkgutil.iter_modules():
    [file_finder, module_name, ispkg] = module_name
    if not ispkg:
        continue
    # print(module_name)
    try:
        module = importlib.import_module(module_name)
        module_list.append([file_finder, module_name, ispkg])
    except:
        pass
    # print(module_name['ispkg'])
    # print(module_name[2])
    # for key in module_name:
    #     print('\t', key)
```

```python
module_name_list = []
def get_module_name(module_obj):
    str_module_obj = str(module_obj)
    # _overlapped <module '_overlapped' from 'D:\\ProgramFile\\Anaconda3\\envs\\dev\\DLLs\\_overlapped.pyd'> False
    add_module_obj = str_module_obj.split("<module '")[1].split("' from ")[0]
    return_bool = True
    if not add_module_obj in module_name_list:
        module_name_list.append(add_module_obj)
        return_bool = False
    return return_bool

class_name_list = []
def get_class_name(class_obj):
    str_class_obj = str(class_obj)
    # <class 'collections.abc.AsyncGenerator'>
    add_class_obj = str_class_obj.split("<class '")[1].split("'")[0]
    return_bool = True
    if not add_class_obj in class_name_list:
        class_name_list.append(add_class_obj)
        return_bool = False
    return return_bool
```

```python
# inspect.getmodule(module)
# for x in inspect.getmodule(module):
#     print(x)
def check_builtin_function(obj_name, obj):
    return_bool = False
    if inspect.isbuiltin(obj):
        # print('inspect.isbuiltin(obj)')
        return_bool = True
    elif str(obj_name) in builtin_modules:
        # print('obj_name in builtin_modules')
        # print(builtin_modules)
        return_bool = True
    elif str(obj).find('(built-in)') > -1:
        # print("str(obj).find('(built-in)')")
        return_bool = True
    elif str(obj_name)[:2] == '__' and str(obj_name)[-2:] == '__':
        return_bool = True
    # elif module_path_change(str(obj)).find(module_base_path) > -1:
    #     return_bool = True
    # print(obj_name[:2], obj_name[-2:])
    # print(str(obj_name) in builtin_modules, obj_name)
    # print('\t\t\t\t', obj_name, obj, return_bool)
    return return_bool
            
def get_modules(module_obj):
    for obj_name, obj in inspect.getmembers(module_obj, inspect.ismodule):
        try:
            # print('\t', obj_name, obj)
            if check_builtin_function(obj_name, obj):
                continue
            if get_module_name(obj):
                continue
            # print('\t', obj_name, obj)
            get_modules(obj)
            get_classes(obj)
            # get_functions(obj)
        except:
            pass
        
def get_classes(module_obj):
    for obj_name, obj in inspect.getmembers(module_obj, inspect.isclass):
        try:
            # print('\t\t', obj_name, obj)
            if check_builtin_function(obj_name, obj):
                continue
            if get_class_name(obj):
                continue
            # print('\t\t', obj_name, obj)
            # get_modules(obj)
            get_classes(obj)
            # get_functions(obj)
        except:
            pass
        
def get_functions(module_obj):
    for obj_name, obj in inspect.getmembers(module_obj, inspect.isfunction):
        try:
            # print('\t\t\t', obj_name, obj)
            if check_builtin_function(obj_name, obj):
                continue
            print('\t\t\t', obj_name, obj)
            # get_modules(obj)
            # get_classes(obj)
            # get_functions(obj)
            # print('\t\t', obj_name, obj)
        except:
            pass
# for x in inspect.getmembers(module, inspect.ismodule):
#     print('*'*25)
#     print(x)

# get_modules(module)
# get_classes(module)
# get_functions(module)
```
